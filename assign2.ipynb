```python
documents = [
    "NLP is exciting and full of possibilities.",
    "Machine learning makes NLP even more powerful.",
    "Text processing is an important part of NLP.",
]
```

```python
tokenized_docs = [word_tokenize(doc.lower()) for doc in documents]
tokenized_docs
```
