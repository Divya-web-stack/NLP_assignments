```python
text = "Hey! I’m enjoying NLP—tokenization is fun, isn't it? Let's try multiple methods."
```

```python
tokens_whitespace = text.split()
tokens_whitespace
```

```python
from nltk.stem import PorterStemmer

porter = PorterStemmer()
porter_stems = [porter.stem(word) for word in tokens_whitespace]
porter_stems
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
lemmatized = [lemmatizer.lemmatize(word) for word in tokens_whitespace]
lemmatized
```
